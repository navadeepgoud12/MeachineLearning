```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
# basic information about dataset
dataset.info()
```

DATA CLEANING

```python
## cleaning the dataset
dataset[dataset.isnull().any(axis=1)]
```

the dataset is converted into two regions based on the index 122, we can make the one column based on the Region

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
```

```python
dataset.info()
```

```python
dataset['Region'] = dataset['Region'].astype(int)
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
df = dataset
```

```python
df.isnull().sum()
```

```python
## removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## removing the spaces from column names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

#### converting the object type to integer type

```python
## convert the object to integers which are integers in nature
df.columns
```

```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
## converting the other features to float
objects = [features for features in df.columns if df.dtypes[features]=='O']
```

```python
objects
```

```python
for feature in objects:
    if feature != 'Classes':
        df[feature] = df[feature].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

#### EDA

```python
## save the cleaned data set
df.to_csv('Algerian_forest_fires_cleaned.csv',index=False)
```

```python
## drop day mount and year and stored in df_copy
df_copy = df.drop(['day','month','year'],axis = 1)
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
# converting the cotegorical column 'Classes' into numerical values
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
# plot density plot for all features
plt.style.use('seaborn-v0_8')  # instead of 'seaborn'
df_copy.hist(bins=30,figsize=(20,15))
plt.show()
```

```python
## creating a pie chart for the fire and not fire instances
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
percentage
```

```python
lablels = ['fire','not fire']
plt.figure(figsize=(10,10))
plt.pie(percentage,labels=lablels,autopct='%1.1f%%')
plt.show()
```

```python
## monthly distribution of fire cases
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
dftemp = df.loc[df['Region']==1]
plt.subplots(figsize=(15,8))
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.xlabel('Month')
plt.ylabel('Number of cases')
plt.title('Monthly Distribution of Fire Cases in Region 1')
plt.legend()
plt.show()
```

```python
dftemp = df.loc[df['Region']==0]
plt.subplots(figsize=(15,8))
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.xlabel('Month')
plt.ylabel('Number of cases')
plt.title('Monthly Distribution of Fire Cases in Region 0')
plt.legend()
plt.show()
```
